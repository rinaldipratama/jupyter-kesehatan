```python
import pandas as pd
import plotly.graph_objs as go

data = pd.read_excel('data/data_kesehatan.xlsx')

# Sort the DataFrame by 'Age'
data = data.sort_values(by='Age')

x = data['Age']
y = data['BodyTemp(fahrenheit)']
risk_levels = data['RiskLevel']

# Define color mappings for risk levels
color_mapping = {'low risk': 'green', 'mid risk': 'yellow', 'high risk': 'red'}

fig = go.Figure()

# Add the line chart trace
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Body Temperature (Â°F)'))

# Add RiskLevel as hover text and set marker color based on RiskLevel
for i in range(len(x)):
    risk_level = risk_levels[i]
    marker_color = color_mapping.get(risk_level, 'gray')  # Default to gray if not in the mapping

    fig.add_trace(go.Scatter(
        x=[x[i]],
        y=[y[i]],
        mode='markers',
        marker=dict(color=marker_color, size=10),
        name='',
        showlegend=False,
        hovertext=f"Age: {x[i]}, Temperature: {y[i]}Â°F, Risk Level: {risk_level}"
    ))

fig.update_layout(
    title='Body Temperature over Age with Risk Level',
    xaxis_title='Age',
    yaxis_title='Body Temperature (Â°F)',
    width=1200,
    height=600
)

fig.show()
```

```python
import pandas as pd
import plotly.graph_objs as go

data = pd.read_excel('data/data_kesehatan.xlsx')

# Sort the DataFrame by 'Age'
data = data.sort_values(by='Age')

x = data['Age']
systolic_bp = data['SystolicBP']
diastolic_bp = data['DiastolicBP']

fig = go.Figure()

# Add the 'SystolicBP' line
fig.add_trace(go.Scatter(x=x, y=systolic_bp, mode='lines', name='SystolicBP'))

# Add the 'DiastolicBP' line with a different line style
fig.add_trace(go.Scatter(x=x, y=diastolic_bp, mode='lines', name='DiastolicBP'))

fig.update_layout(
    title='Systolic and Diastolic Blood Pressure over Age',
    xaxis_title='Age',
    yaxis_title='Blood Pressure',
    width=1200,
    height=600
)

fig.show()
```

```python
import pandas as pd
import plotly.graph_objs as go

data = pd.read_excel('data/data_kesehatan.xlsx')

# Sort the DataFrame by 'Age'
data = data.sort_values(by='Age')

x = data['Age']
systolic_bp = data['SystolicBP']
diastolic_bp = data['DiastolicBP']

fig = go.Figure()

# Add 'SystolicBP' as a bar chart
fig.add_trace(go.Bar(x=x, y=systolic_bp, name='SystolicBP', marker_color='blue'))

# Add 'DiastolicBP' as a bar chart
fig.add_trace(go.Bar(x=x, y=diastolic_bp, name='DiastolicBP', marker_color='green'))

fig.update_layout(
    title='Systolic and Diastolic Blood Pressure over Age (Bar Chart)',
    xaxis_title='Age',
    yaxis_title='Blood Pressure',
    width=1200,
    height=600
)

fig.show()
```
